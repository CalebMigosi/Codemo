# Installing packages

## What are packages?
Packages are simply folders containing python code that has been optimized and posted on the internet for all users to benefit from.

The common packages are pandas, numpy, matplotlib and scikit-learn.

Before being able to use packages,we have to install them into our computers like we would for an online program.

Once installed, we have to add them to our Python code. We do this by ***importing*** the packages into the script.

### Installing packages from the internet
Take as an example we want to install the `pandas` package.

To install packages, you can go directly to the ***Anaconda Prompt*** if you are on Windows or press Cmd + Space and type "Terminal".

You can then type: `pip install pandas`

Similarly for `numpy` you type `pip install numpy`


### Installing packages directly on Jupyter Notebook
You have a small shortcut for installing packages directly on Jupyter. You simply add an exclamation mark before writing `pip install pandas`.

### Importing Packages

To import packages, you can simply type `import pandas`.

If you only want to import a small part of the package you can type `from pandas import ` and the small section of of the package you want to install.

Do not worry if you don't know any/all the small sub parts of a package. We're lucky to have the internet now and any time you feel stuck, always use the internet. I will show how to do this in the next segment of this class.

```python
# These are different ways of importing a package
import investpy
from investpy import crypto
import investpy.crypto as cpto
```

```python
# Using each of the methods above, we've used the package to get bitcoin prices from 2018 to 2021 January
dataset1 = investpy.crypto.get_crypto_historical_data(crypto='bitcoin', 
                                           from_date='01/01/2018', 
                                           to_date='01/01/2021')

dataset2 = crypto.get_crypto_historical_data(crypto='bitcoin', 
                                           from_date='01/01/2018', 
                                           to_date='01/01/2021')

dataset3 = cpto.get_crypto_historical_data(crypto='bitcoin', 
                                           from_date='01/01/2018', 
                                           to_date='01/01/2021')
```

```python
# We can test to see that the data is essentially the same
dataset1 == dataset2 
```
